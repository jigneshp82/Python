```python
logs = pd.read_csv('/home/jignesh/Documents/Python/IN1469_student_files_20190920/student_files/resources/new_access.log'
                  ,sep='\s+'
                  ,error_bad_lines=False
                  ,header=None
                   ,usecols=(0,3,5,6,7,8)
                ,names = ['addr','request_date','reqest','status','size','brozer'])
```

```python
logs.head(10)
```

```python
print(logs.info())

logs.request_date= pd.to_datetime(logs.request_date,format = '[%d/%b/%Y:%H:%M:%S')

print(logs.info())
```

```python
top10addr = logs.groupby("addr").size().sort_values(ascending = False).head(20)

print(top10addr)

top10addr[::-1].plot(kind='barh')
```
